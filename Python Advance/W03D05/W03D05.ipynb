```python
## we can save the state of the fun.

def even_generator_fun():
    n = 0
    yield n
    n = n+2
    yield n
    n = n+2
    yield n
    n = n+2
    yield n
```

```python
iter_obj = even_generator_fun()
```

```python
next(iter_obj)
```

```python
next(iter_obj)
```

```python
next(iter_obj)
```

```python
next(iter_obj)
```

```python
next(iter_obj)
```

```python
## Create an infinite seq.

def infinite_even_gen_fun():
    n=0
    while True:
        yield n
        n = n+2
iter_obj = infinite_even_gen_fun()
```

```python
next(iter_obj)
```

```python
next(iter_obj)
```

```python
next(iter_obj)
```

```python
for i in range(30):
    print(next(iter_obj))
```

```python
print(next(iter_obj))
```
