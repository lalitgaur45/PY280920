```python
1. Write a function which takes a string
2. Check if the string contains a vowel
3. If it does, print the string once
4. else print the string n times(n should be passed by the user when calling the fun)

Try to break the code in few functions
```

```python
def checkifvowels(string):
    vowels = "aeiou"
    for ch in string:
        if ch in vowels:
            return True
    else:
        return False
```

```python
checkifvowels("apple")
```

```python
checkifvowels("ppl")
```

```python
def print_string_n_times(string,n):
    for i in range(n):
        print(string)
```

```python
print_string_n_times("apple",4)
```

```python
def task(string,n):
    check = checkifvowels(string)
    if check == True:
        print(string)
    else:
        print_string_n_times(string,n)
```

```python
task("apple",4)
```

```python
task("ppl",4)
```

```python
def task(**kwargs):
    print(type(kwargs))
    print(kwargs['name'],"is studying at",kwargs['college'],"this guy took",kwargs['stream'])
```

```python
task(college="MIT",name="rohan rathore",stream="Computer Science")
```

```python
def task(*args):
    print(type(args))
    print(args[1],"is studying at",args[0],"this guy took",args[2])
```
